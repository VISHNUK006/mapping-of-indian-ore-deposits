### <center style="text-align:center;background-color: #D7E6AC"><u>Introduction to mining in India</u></center>
    
> <p>The mining industry in India serves as a cornerstone of economic activity, significantly bolstering the nation’s growth and industrial development. While the direct contribution of mining to India’s GDP ranges from 2.2% to 2.5%, its impact on the industrial sector is far more substantial, accounting for approximately 10% to 11%. Notably, even small-scale mining operations contribute around 6% to the overall mineral production costs. The industry provides employment opportunities to nearly 700,000 individuals, underscoring its role in enhancing livelihoods across the country.</p>
>    
> <p>As of recent data, India stands as a leading global player in mineral production, being the largest producer of sheet mica and the fourth largest producer of iron ore, alumina, chromite, and bauxite as of 2015. Additionally, India boasts the fifth-largest reserves of coal and iron ore worldwide, showcasing its vast resource potential. The Indian metal and mining industry was estimated to be valued at $106.4 billion in 2010, reflecting its critical importance to the national economy.</p>
> 
> <p>In 2019, India ranked as the fourth-largest producer of iron ore and chromium globally, the fifth-largest producer of bauxite and zinc, and the seventh-largest producer of manganese and lead. Furthermore, it was the eleventh-largest producer of uranium in 2018, highlighting the diverse array of minerals extracted across the country. </p>
> 
> <p>Mining has deep historical roots in India, dating back to ancient civilizations that recognized the value of mineral resources. While the industry has made significant contributions to the nation's economic landscape, it has also faced challenges, including human rights violations and environmental degradation. Recent high-profile scandals have drawn attention to the urgent need for sustainable and responsible mining practices to mitigate adverse impacts while harnessing India's mineral wealth for future generations.</p>

```python
#function to return the value with its percent
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
```
